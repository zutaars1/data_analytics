# Exploratory data analysis (EDA) of Supermarkets data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read and select variables
df = pd.read_csv("supermarkets_data_enriched.csv")[['id',
                                                    'brand',
                                                    'lat',
                                                    'lon']]

# Subset
df_sub = df.loc[df['brand'].isin(['Coop', 'Migros', 'Landi', 'Denner', 'Volg'])]
df_sub

# Colors
colors = {'Coop': 'green', 'Migros': 'orange', 'Landi': 'blue', 'Denner': 'red', 'Volg': 'purple'}

# Plot
plt.scatter(df_sub['lon'],
            df_sub['lat'],
            c=df_sub['brand'].map(colors))
```
